```python
import numpy as np
import pandas as pd 
import operator 

ϕ = np.random.rand(40,10) 
ϕ 
```

```python
ψ = np.random.rand(40) 
ψ 
```

```python
least_squares_solution, loss, rank, s = np.linalg.lstsq(ϕ, ψ, rcond=None)
```

```python
least_squares_solution
```

```python
associated_loss = np.linalg.norm(np.dot(ϕ,least_squares_solution) - ψ)
associated_loss
```

```python
δ = np.random.rand(10) 
```

```python
(np.dot(ϕ , (least_squares_solution + δ))).shape
```

```python
lhs_eq = np.dot(ϕ , (least_squares_solution + δ)) - ψ
print(np.linalg.norm(lhs_eq))
```

```python
lhs = np.linalg.norm(lhs_eq) ** 2
lhs
```

```python
rhs = np.linalg.norm(associated_loss) ** 2 
rhs
```
